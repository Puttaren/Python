Uppgift 10.2.3

```python
# Verifiera att AB = AC, men att B != C (det senare ser vi visuellt)
A = np.array([
    [1,2],
    [2,4]    
])

B = np.array([
    [2,1],
    [1,3]
])

C = np.array([
    [4,3],
    [0,2]
])

AB = A @ B

AC = A @ C

print (AB)
print ()
print (AC)
print ()
if np.array_equal(AB, AC):
    print("Matriserna är identiska trots att B och C inte är det")
print()

```
